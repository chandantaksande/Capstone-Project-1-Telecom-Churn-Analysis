# 6.3 Dataset Features  

1. State: It contains name of states (Total 51 states)

2. Account Length: It displays the number of months the account has been active.

3. Area Code: These are the codes for the specific areas that are used to identify them.

4. International Plan: Yes-indicates that the customer had International Plan, while No-indicates that the customer had not subscribed to International Plan.

5. Voice Mail Plan: Yes indicates that the customer had Voice Mail Plan, while No indicates that the customer had not subscribed to Voice Mail Plan.

6. Number vmail messages: Number of voice messages done by customers.

7. Total day minutes: Total amount of time spent in the morning.

8. Total day calls: Total number of calls made by the customer in the morning.

9. Total day charge: It is the total cost of services used by customers in the morning.

10. Total eve minutes: Total amount of time spent in the Evening.

11. Total eve calls: Total number of calls made by the customer in the Evening.

12. Total eve charge: It is the total cost of services used by customers in the evening.

13. Total night minutes: Total amount of time spent in the Night.

14. Total night calls: Total number of calls made by the customer in the Night.

15. Total night charge: It is the total cost of services used by customers in the Night.

16. Total intl minutes: Total amount of time spent by customer over international calls.

17. Total intl calls: Total of the number of international calls made by the customer

18. Total intl charge: It is the total cost of international services used by customers

19. Customer service calls: Number of calls made by customer to the customer service

20. Churn: This is our dependent variable that describes whether or not customers is churned.

# 7.3 Account Length Vs Churn

Account Length refers to from how long user has been taking services of our company. So lets see if users duration with our services has any impact on churn rate.We divided account length into three categories.
1. From 0 to 12 (Short Term Users)
2. From 13 to 100 (Middle Term Users)
3. From 100 onwards (Long Term Users)


```python
# Finding short term user which are using using service from not long i.e. from 0 to 12
short_term_user=telecom_churn_df[(telecom_churn_df['Account length']>0) & (telecom_churn_df['Account length']<=12)].loc[:,['Churn']].value_counts()
short_term_user
```

```python
# Ploting pie chart for short_term_user
plt.figure(figsize=(10,6))
plt.pie(short_term_user, explode = (0,0.3),autopct='%1.1f%%',shadow=True,radius = 2.0, labels = ['Not churned customer','Churned customer'])
circle = plt.Circle( (0,0), 1, color='white')
p=plt.gcf()
plt.title("Short Term User vs Churn")
p.gca().add_artist(circle)
plt.show()
```

```python
# Finding middle term user which are using using service from 13 to 100
middle_term_user=telecom_churn_df[(telecom_churn_df['Account length']>=13) & (telecom_churn_df['Account length']<=100)].loc[:,['Churn']].value_counts()
middle_term_user
```

```python
# ploting donought chart for middle_term_user
plt.figure(figsize=(10,6))
plt.pie(middle_term_user, explode = (0,0.3),autopct='%1.1f%%',shadow=True,radius = 2.0, labels = ['Not churned customer','Churned customer'])
circle = plt.Circle( (0,0), 1, color='white')
p=plt.gcf()
plt.title("Middle Term User vs Churn")
p.gca().add_artist(circle)
plt.show()
```

```python
# Finding long term user which are using using service from long period i.e. from 101 to 243
long_term_users=telecom_churn_df[(telecom_churn_df['Account length']>100) & (telecom_churn_df['Account length']<=telecom_churn_df['Account length'].max())].loc[:,['Churn']].value_counts()
long_term_users
```

```python
plt.figure(figsize=(10,6))
plt.pie(long_term_users, explode = (0,0.3),autopct='%1.1f%%',shadow=True,radius = 2.0, labels = ['Not churned customer','Churned customer'])
circle = plt.Circle( (0,0), 1, color='white')
p=plt.gcf()
plt.title("Long Term User vs Churn")
p.gca().add_artist(circle)
plt.show()
```

Account length i.e. time period from which user is using services, has no effect on churn rate whatsoever

#7.7 Calling Time VS Churn

Here we wanted to see if total calling time has some kind of effect on churn rate.
so we added new column called "Total Talk time" and plot it against churn.

Note:- In "Total Talk Time" column value, we didn't add 'Total intl minutes' in 'Total Talk Time' cause we already did analysis for international users above. And it already has big impact on churn users.If added it would have affected the result and We also wanted to find out other factors which are affecting churn rate 

```python
# Creating new column called "Total Talk Time" by adding "Total Day Minutes","Total eve Minutes", "Total night minutes" together
telecom_churn_df["Total Talk Time"]=telecom_churn_df["Total day minutes"]+telecom_churn_df["Total eve minutes"]+telecom_churn_df["Total night minutes"]
telecom_churn_df
```

```python
# Checking relation between "Total Talk Time" vs "Churn" in Tabular form
pd.crosstab(telecom_churn_df["Churn"],telecom_churn_df["Total Talk Time"])
```

```python
# Ploting above obtained value as it in countplot
plt.figure(figsize=(15,7.5))
sns.countplot(x="Total Talk Time",data=telecom_churn_df,hue="Churn",palette = "Set2")
plt.title("Total Talk Time")
plt.xlabel("Total Talk Time")
plt.ylabel("Avg. User")
```

As the talktime along the x-axis increases, churn users(Red Lines) also increases. And at the end of x axis we can see that red lines i.e. churned users are more.

```python
# Ploting boxplot for better understanding and visualization of data
sns.boxplot(x='Churn', y='Total Talk Time', data = telecom_churn_df)
plt.show()
```

It is observe that as total talk time increases, so does the churn rate.

#7.10 Customer service calls vs Churn

```python
# Finding relational data between "Customer service calls" and "Churn" data for better understanding
telecom_churn_df.groupby(["Customer service calls"])["Churn"].value_counts().unstack()
```

```python
#Ploting countplot for above obtain data for better visualization 
plt.figure(figsize=(15,7.5))
sns.countplot(x="Customer service calls",data=telecom_churn_df,hue="Churn",palette = "Set2")
plt.title("Customer service calls")
plt.xlabel("Customer service calls")
plt.ylabel("Avg. User")
```

```python
# Creating a column which has record of more than 3 service calls only
telecom_churn_df['High Service call'] = telecom_churn_df['Customer service calls'] > 3

# Finding relational data between "High service calls" and "Churn" data and storing it in new variable
cross_table_high_service_calls_vs_churn=pd.crosstab(telecom_churn_df['High Service call'], telecom_churn_df['Churn'])

#printing whole cross table present in "cross_table_high_service_calls_vs_churn"
print("High Service calls vs churn users and non-churn users")
print(cross_table_high_service_calls_vs_churn)

#printing only high service call users data from "cross_table_high_service_calls_vs_churn" 
print("\n\nHigh Service calls vs only churn users")
print(cross_table_high_service_calls_vs_churn.iloc[:,1:])
```

```python
#Ploting donought chart for only high service users against churn
plt.figure(figsize=(10,6))
plt.pie(cross_table_high_service_calls_vs_churn.iloc[:,1:], explode = (0,0.3),autopct='%1.2f%%',shadow=True,radius = 2.0, labels = ['Not churned customer','Churned customer'])
circle = plt.Circle( (0,0), 1, color='white')
p=plt.gcf()
p.gca().add_artist(circle)
plt.show()
```

Users who are making more than 3 customer service calls, also has very high churn rate of 28.57% (Refer 7.10). User generally do customer service calls for complaints related to network and service issues. And we think churn rate is so high, because unresolved complaints of users.
